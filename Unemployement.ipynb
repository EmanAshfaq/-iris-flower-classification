```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Pakistan_Poverty_Dataset_2000_2023.csv")

# Show first 5 rows
df.head()
```

```python
print(df.columns.tolist())
```

```python
print(df["Unemployment Rate (%)"].unique())
```

```python
df["Unemployment Rate (%)"] = df["Unemployment Rate (%)"].astype(str).str.replace("%", "").str.strip()
df["Unemployment Rate (%)"] = pd.to_numeric(df["Unemployment Rate (%)"], errors="coerce")
df.rename(columns={"Unemployment Rate (%)": "unemployment_rate"}, inplace=True)
```

```python
df["year"] = pd.to_datetime(df["Year"], errors="coerce").dt.year
df.dropna(subset=["year", "unemployment_rate"], inplace=True)
df["year"] = df["year"].astype(int)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="year", y="unemployment_rate", marker="o", color="darkgreen")
plt.title("Unemployment Rate in Pakistan (20002023)")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
